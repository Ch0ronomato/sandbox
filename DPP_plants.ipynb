```python
import sys, os, re, itertools
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy import linalg
from sklearn import mixture
from sklearn import manifold
from sklearn.metrics import euclidean_distances
from sklearn.decomposition import PCA
%pylab inline
plt.rcParams['figure.figsize'] = 7, 7
```

```python
def train_dppmix(blen_matrix):
    # train classifiers
    classifiers = {}
    for cov_mat in ['tied', 'full', 'diag', 'spherical']:
        classifiers[cov_mat] = mixture.DPGMM(n_components = 60, n_iter = 1000, covariance_type=cov_mat)
        classifiers[cov_mat].fit(blen_matrix)
    
    # find best according to the AIC
    bic_models = [[], []]
    bic_models[0] = ['tied', 'full', 'diag', 'spherical']
    bic_models[1] = [classifiers[m].bic(blen_matrix) for m in classifiers]
    bic_models = np.array(bic_models).transpose()
    bic_models = bic_models[bic_models[:, 1].argsort(),:]
    best_model = bic_models[0, 0]
    
    mds = manifold.MDS(n_components = 3, max_iter = 3000, dissimilarity='precomputed')
    pos = mds.fit(euclidean_distances(angio_mat)).embedding_ 
    
    #first item in the tupple is the position of the points in MDS, while
    #the second is the bic scores of the different covariance matrices, and the 
    #third is the classifier with highest bic score
    return (pos, bic_models, classifiers[best_model])
```

## Angiosperm data set

 - It has 76 genes and 18 taxa. Other notes?

```python
angio_mat = np.array(pd.read_csv('angiosperm_matrix.csv'))
angio_dpp = train_dppmix(angio_mat)
print 'The number of clusters is: %s' %len(set(angio_dpp[2].predict(angio_mat)))
print angio_dpp[1]
print pd.Series(angio_dpp[2].predict(angio_mat)).value_counts()[:5]
```

```python
color_iter = itertools.cycle(['r', 'g', 'b', 'c', 'm'])
clf = angio_dpp[2]
pos = angio_dpp[0]
Y_ = clf.predict(angio_mat)
#splot = plt.subplot(1, 1, 1)
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
for i, (mean, covar, color) in enumerate(zip(clf.means_, clf._get_covars(), color_iter)):
    v, w = linalg.eigh(covar)
    u = w[0] / linalg.norm(w[0])
    if not np.any(Y_ == i):
        continue
    ax.scatter(pos[Y_ == i, 0], pos[Y_==i, 1], pos[Y_==i, 2],color = color)
```
